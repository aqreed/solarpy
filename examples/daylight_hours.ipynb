```python
import numpy as np
import matplotlib.pyplot as plt
import solarpy as sp
from datetime import datetime, timedelta
```

Create the mesh grid for the "day of the year" and the latitude:

```python
n = np.array([datetime(datetime.now().year, 1, 1, 6, 0) +
               timedelta(days=i) for i in range(365)])
lat = np.linspace(-90, 90, 300)

N, LAT = np.meshgrid(n, lat)
```

To apply the function *daylight_hours* to the mesh, it is important to use **np.vectorize**:

```python
light_hours = np.vectorize(sp.daylight_hours)(N, LAT)
```

Then visualize as a 2D filled contour plot:

```python
plt.figure(figsize=(18, 6))
plt.contourf(n, lat, light_hours, [0, 1, 6, 9, 10, 11, 12,
                                     13, 14, 15, 17, 23, 24])
plt.colorbar(label='nÂº of hours')
plt.xlabel('Month', fontsize=15)
plt.ylabel('latitude', fontsize=15)
plt.title('Daylight hours', fontsize=18)
plt.show()
```
